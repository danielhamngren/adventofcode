```python
with open('input') as f:
    lines = f.read().splitlines()
```

# Part 1

```python
int_lines = [int(x) for x in lines]
sum(int_lines)
```

# Part 2

```python
current_frequency = 0

historical_frequencies = {current_frequency}

found_frequency_twice = False

i = 0

# int_lines = [1, -1]
# int_lines = [+3, +3, +4, -2, -4]
# int_lines = [-6, +3, +8, +5, -6]
# int_lines = [+7, +7, -2, -7, -4]

while not found_frequency_twice:
    if i >= len(int_lines):
        i = 0
        print('i = 0', current_frequency, len(historical_frequencies))
        #print(historical_frequencies)
    
    current_frequency += int_lines[i]
    
    if current_frequency in historical_frequencies:
        print(current_frequency)
        found_frequency_twice = True
    else:    
        historical_frequencies.add(current_frequency)
    i += 1
    
```
