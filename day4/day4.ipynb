```python
with open('input.txt') as f:
    lines = f.read().splitlines()
```

```python
lines = sorted(lines)
display(lines)
```

```python
import re
```

```python
current_shift = [0] * 60
current_guard = None

guards = {}

for line in lines:
    print(line)
    if "shift" in line:
        if current_guard is not None:
            if current_guard in guards:
                guards[current_guard].append(current_shift)
            else:
                guards[current_guard] = [current_shift]
        current_shift = [0] * 60
        numbers = re.findall(r'\d+', line)
        current_guard = numbers[-1]
        print(current_guard)
    else:
        date = re.findall(r'\d+-\d+-\d+', line)
        time = re.findall(r'\d+:\d+', line)
        minute = int(time[0].split(':')[1])
        
        if 'falls asleep' in line:
            current_shift[minute:] = [1] * len(current_shift[minute:-1]) 
        
        elif 'wakes up' in line:
            current_shift[minute:] = [0] * len(current_shift[minute:-1]) 
        print(date, time, current_shift)    
```

```python


max_sleep = -1
max_sleep_guard = None

for guard_id, awake in guards.items():
    total = [sum(x) for x in zip(*awake)]
    #print(guard_id, total)
    if sum(total) > max_sleep:
        max_sleep = sum(total)
        max_sleep_guard = guard_id
    
print(max_sleep_guard)

total_max = [sum(x) for x in zip(*guards[max_sleep_guard])]
print(total_max)

print('Answer 1:', total_max.index(max(total_max)) * int(max_sleep_guard))
```
